```python
from datetime import datetime

numbers1 = [1, 3, 4, 5, 8, 2, 1, 4, 5, 9, 5]
hand1 = "right"
result1 = "LRLLLRLLRRL"

numbers2 = [7, 0, 8, 2, 8, 3, 1, 5, 7, 6, 2]
hand2 = "left"
result2 = "LRLLRRLLLRR"

numbers3 = [1, 2, 3, 4, 5, 6, 7, 8, 9, 0]
hand3 = "right"
result3 = "LLRLLRLLRL"
```

```python
def solution(numbers, hand):
    start_time = datetime.now()
    result=""
    left = (0,0)
    right = (2,0)
    for number in numbers:
        target = 0
        if number == 1:
            target = (0,3)
        elif number == 2:
            target = (1,3)
        elif number == 3:
            target = (2,3)
        elif number == 4:
            target = (0,2)
        elif number == 5:
            target = (1,2)
        elif number == 6:
            target = (2,2)
        elif number == 7:
            target = (0,1)
        elif number == 8:
            target = (1,1)
        elif number == 9:
            target = (2,1)
        elif number == 0:
            target = (1,0)
        
        if (abs(target[0] - left[0]) + abs(target[1]-left[1])) > \
        (abs(target[0] - right[0]) + abs(target[1]-right[1])):
            result+="R"
            right=target
        elif (abs(target[0] - left[0]) + abs(target[1]-left[1])) < \
        (abs(target[0] - right[0]) + abs(target[1]-right[1])):
            result+="L"
            left=target
        elif (abs(target[0] - left[0]) + abs(target[1]-left[1])) == \
        (abs(target[0] - right[0]) + abs(target[1]-right[1])):
            if hand == "right":
                result+="R"
                right = target
            elif hand == "left":
                result+="L"
                left = target
        else:
            print('오류발생')
                
    
    
    answer = result
    
    end_time=datetime.now()
    print(start_time-end_time)
    return answer

print(f'{solution(numbers1, hand1)} == {result1}')
print(f'{solution(numbers2, hand2)} == {result2}')
print(f'{solution(numbers3, hand3)} == {result3}')
```
